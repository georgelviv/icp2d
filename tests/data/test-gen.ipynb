```python
import numpy as np

np.random.seed(42)
base_points = np.random.uniform(-10, 10, (50, 2))  # 50 random 2D points

# Define transformation parameters
rotation_angle = np.radians(5)  # 20-degree rotation
translation_vector = np.array([1, 1])

rotation_matrix = np.array([
  [np.cos(rotation_angle), -np.sin(rotation_angle)],
  [np.sin(rotation_angle), np.cos(rotation_angle)],
])

# Apply transformation to the base points
transformed_points = (base_points @ rotation_matrix.T) + translation_vector

base_points_path = './base-points.csv'
transformed_points_path = './transformed-points.csv'

np.savetxt(base_points_path, base_points, delimiter=",", header="x,y", comments="")
np.savetxt(transformed_points_path, transformed_points, delimiter=",", header="x,y", comments="")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 8))
plt.scatter(base_points[:, 0], base_points[:, 1], label="Base Points", color="blue")
plt.scatter(transformed_points[:, 0], transformed_points[:, 1], label="Transformed Points", color="red", alpha=0.7)
plt.legend()
plt.title("ICP Test Data")
plt.xlabel("X")
plt.ylabel("Y")
plt.axis("equal")
plt.grid()
plt.show()
```
